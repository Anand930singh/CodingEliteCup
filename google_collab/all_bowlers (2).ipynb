```python
def weather(data):
  data['temp'].fillna('60', inplace=True)
  data['humidity'].fillna('65', inplace=True)
  data['windspeed'].fillna('5', inplace=True)
  data['conditions'].fillna('Partially cloudy', inplace=True)

  return data

df=weather(df)
```

```python
columns_to_replace = ['Overs', 'Runs', 'Wkts']
for col in columns_to_replace:
    df[col].replace('-', '0', inplace=True)
```

```python
def convert_player(one_data):
  indices=[]
  for i in range(len(one_data)):
    if one_data['Player'][i] not in indices:
      indices.append(one_data['Player'][i])

  for i in range(len(one_data)):
    one_data['Player'][i]=np.uint8(indices.index(one_data['Player'][i]))

  return one_data

df = convert_player(df)
```

```python
def to_float(data):
  for i in range(len(data)):
    data['Overs'][i]=float(data['Overs'][i])
    data['Runs'][i]=float(data['Runs'][i])
    data['Wkts'][i]=float(data['Wkts'][i])
    data['temp'][i]=float(data['temp'][i])
    data['humidity'][i]=float(data['humidity'][i])
    data['windspeed'][i]=float(data['windspeed'][i])

  return data

to_float(df)
```

```python
classes=[]
for i in range(0,310,10):
  classes.append(i)
```

```python
import math
def change_runs(one_hot_encoded):
  for i in range(len(one_hot_encoded['Runs'])):
    one_hot_encoded['Runs'][i]=np.uint8(classes.index(math.floor(one_hot_encoded['Runs'][i]/10)*10))

  return one_hot_encoded

change_runs(df_encoded)
```

```python
y = pd.Categorical(y, categories=classes)
```

```python
y = pd.get_dummies(y, columns=['Wkts'])
```

```python
from sklearn.model_selection import train_test_split

# Split into training and test sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=X['Player'], random_state=42)

# Further split the test set into validation and final test sets (20% validation, 10% final test)
X_val, X_final_test, y_val, y_final_test = train_test_split(X_test, y_test, test_size=0.33, random_state=42)
```

```python
import numpy as np
X_train=X_train.astype(np.float32)
y_train=y_train.astype(np.float32)
X_val=X_val.astype(np.float32)
y_val=y_val.astype(np.float32)
X_test=X_test.astype(np.float32)
y_test=y_test.astype(np.float32)
```

```python
import tensorflow as tf
```

```python
X_train=tf.convert_to_tensor(X_train)
y_train=tf.convert_to_tensor(y_train)
X_val=tf.convert_to_tensor(X_val)
y_val=tf.convert_to_tensor(y_val)
X_test=tf.convert_to_tensor(X_test)
y_test=tf.convert_to_tensor(y_test)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import regularizers
from keras.optimizers import Adam
```

```python
model= keras.Sequential([
    keras.layers.Dense(512,activation="relu"),
    keras.layers.Dense(512,activation="relu"),
    keras.layers.Dense(512,activation="relu"),
    keras.layers.Dense(256,activation="relu"),
    keras.layers.Dense(256,activation="relu"),
    keras.layers.Dense(128,activation="relu"),
    keras.layers.Dense(64,activation="relu"),
    keras.layers.Dense(31,activation="softmax")
])
```

```python
optimizer=Adam(learning_rate=0.00001)
```

```python
model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=10,batch_size=64,validation_data=(X_val,y_val))
```

```python
test_loss,test_acc=model.evaluate(X_test,y_test)
print("tested accuracy :",test_acc*100)
```
